
# 🔐 Secret Scopes

---

## 1️⃣ Create Scope & Store Secrets

Secret Scopes can be managed using:

- **Azure Key Vault-backed**
- **Databricks-backed**

---

### A) Azure Key Vault-backed (via Databricks UI)
*(Secrets stored in Azure Key Vault)*

- **Docs:** https://learn.microsoft.com/en-us/azure/databricks/security/secrets/
- **Flow:** `Azure Key Vault → Databricks Secret Scope → Notebooks / Clusters / Jobs`

#### Azure

Create an **Azure Key Vault** and add secrets under:

```
Objects
└── Secrets
```

Example secrets:

- `ADLSAccessKeys`
- `Tokens`

#### Databricks

Open:

```text
https://<databricks-url>#secrets/createScope
```

Create the Secret Scope:

| Property | Value |
|----------|-------|
| Scope Name | `customscopename` |
| DNS Name | `<Vault URI from Key Vault Properties>` |
| Resource ID | `<Key Vault Resource ID from Properties>` |

Access the secret:

```python
dbutils.secrets.get(scope="<scope-name>", key="<secret-key>")
```

---

### B) Databricks-backed (via Python SDK)
*(Secrets stored in Databricks)*

```python
from databricks.sdk import WorkspaceClient

w = WorkspaceClient()

w.secrets.create_scope(scope=secret_scope_name)                                  # Create scope
w.secrets.delete_scope(scope=secret_scope_name)                                  # Delete scope

w.secrets.put_secret(
    scope=secret_scope_name,
    key=secret_key_name,
    string_value=connection_json
)                                                                                # Store secret

w.secrets.delete_secret(
    scope=secret_scope_name,
    key=secret_key_name
)                                                                                # Delete secret
```

---

### C) Databricks-backed (via Databricks CLI)
*(Secrets stored in Databricks)*

```bash
databricks secrets create-scope <scope-name>
databricks secrets delete-scope <scope-name>

databricks secrets put-secret <scope-name> <secret-key> "value_string"
databricks secrets delete-secret <scope-name> <secret-key> "value_string"

databricks secrets list-scopes
databricks secrets list-secrets <scope-name>
```

## 2️⃣ Verify Scopes, Keys & Retrieve Secrets

List all Secret Scopes:

```python
dbutils.secrets.listScopes()
```

List all keys in a Secret Scope:

```python
dbutils.secrets.list(scope="azdbxsecretscope")
```

Retrieve a secret:

```python
my_secret = dbutils.secrets.get(
    scope="azdbxsecretscope",
    key="ADLSAccessKeys"
)
```

> **Note:** Databricks automatically redacts secret values in notebook outputs.

````python
from databricks.sdk import WorkspaceClient

w = WorkspaceClient()

# Scope name
secret_scope_name = "restaurantproject"

# Create Secret Scope
w.secrets.create_scope(scope=secret_scope_name)

# Store Secrets
w.secrets.put_secret(
    scope=secret_scope_name,
    key="EVENTHUB_NAME",
    string_value=<eventhub-name>
)

w.secrets.put_secret(
    scope=secret_scope_name,
    key="EVENTHUB_NAMESPACE",
    string_value=<eventhub-namespace>
)

w.secrets.put_secret(
    scope=secret_scope_name,
    key="EVENTHUB_CONN_STR",
    string_value=<eventhub-conn-string>
)
````
